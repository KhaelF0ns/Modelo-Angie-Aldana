```python
nlp = spacy.blank("en")
url_pattern = re.compile(r"https?://[^,; ]+")
char_pattern = re.compile(r"[^a-z ]")
spaces_pattern = re.compile(r"\s+")
parts_pattern = re.compile(r"\[[^]]*\]")
noIntro_pattern = re.compile(r"^.*Lyrics")
noFinal_pattern = re.compile(r"\d*embed")
```

```python
def preprocess(text: str) -> str:
	try:
		noIntro = re.sub(noIntro_pattern, " ", text)
		lower_text = str(noIntro).lower()
		no_parts = re.sub(parts_pattern, " ", lower_text)
		no_double_spaces = re.sub(spaces_pattern, " ", no_parts)
		noFinal = re.sub(noFinal_pattern, "", no_double_spaces)
		no_special = re.sub(char_pattern, "", noFinal)

		doc = nlp(no_special)
		valid_text = " ".join(
			map(
				lambda token: token.text,
				filter(lambda token: (
					not token.is_stop and
					len(token.text) < 24 and
					len(token.text) > 2
				), doc)
			)
		)
		return valid_text.strip()
	except:
		return None
```

```python
print(preprocess(df.Lyrics.iloc[1]))
```

```python
df["Lyrics_proces"] = df.Lyrics.apply(preprocess)
```
